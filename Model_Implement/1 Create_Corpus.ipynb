```python
# Cell: Load Thai corpus
from datasets import load_dataset, concatenate_datasets

print("ðŸ“¦ Loading Thai corpus...")

# à¹‚à¸«à¸¥à¸” datasets à¸•à¸²à¸¡à¸—à¸µà¹ˆ doc à¸šà¸­à¸
wiki_ds = load_dataset("pythainlp/thai-wiki-dataset-v3", split="train")
thaigov_ds = load_dataset("pythainlp/thaigov-corpus", split="train")
wikitext_ds = load_dataset("Salesforce/wikitext", "wikitext-103-v1", split="train")

print(f"âœ… Thai Wiki: {len(wiki_ds)} samples")
print(f"âœ… Thai Gov: {len(thaigov_ds)} samples")
print(f"âœ… Wikitext: {len(wikitext_ds)} samples")

# à¸£à¸§à¸¡ corpus
def extract_text(example):
    # à¸”à¸¶à¸‡ text field (à¹à¸•à¹ˆà¸¥à¸° dataset à¸­à¸²à¸ˆà¸¡à¸µà¸Šà¸·à¹ˆà¸­à¸•à¹ˆà¸²à¸‡à¸à¸±à¸™)
    if 'text' in example:
        return {'text': example['text']}
    elif 'content' in example:
        return {'text': example['content']}
    else:
        return {'text': str(example)}

# Combine all
all_texts = []

print("\nðŸ”¨ Extracting texts...")
for item in wiki_ds:
    if 'text' in item and len(item['text']) > 50:
        all_texts.append(item['text'])

for item in thaigov_ds:
    if 'text' in item and len(item['text']) > 50:
        all_texts.append(item['text'])

for item in wikitext_ds:
    if 'text' in item and len(item['text']) > 50:
        all_texts.append(item['text'])

print(f"âœ… Total texts: {len(all_texts)}")

# Save to file
with open('thai_corpus.txt', 'w', encoding='utf-8') as f:
    f.write('\n\n'.join(all_texts))

print(f"âœ… Saved to thai_corpus.txt")
```
